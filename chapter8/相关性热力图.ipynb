```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
classdata = pd.read_csv('class.csv')
classdata.head()
```

```python
classdata.plot.scatter(x='Age', y='Height', s=80, c='g')

plt.show()
```

```python
classdata.plot.scatter(x='Age', y='Height', s=80, c='g')

plt.figure()
plt.scatter(classdata['Height'], classdata['Weight'], c='b')
plt.xlabel('身高')
plt.ylabel('体重')
plt.show()

corr = classdata.corr()

# 热力图
plt.figure()
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns, yticklabels=corr.columns)

plt.show()

corr
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 示例数据读取（假设classdata已经存在）

# 绘制散点图
classdata.plot.scatter(x='Age', y='Height', s=80, c='g')

plt.figure()
plt.scatter(classdata['Height'], classdata['Weight'], c='b')
plt.xlabel('身高')
plt.ylabel('体重')
plt.show()

# 只选择数值列用于计算相关系数
numeric_data = classdata.select_dtypes(include=['float64', 'int64'])

# 计算相关系数
corr = numeric_data.corr()

# 热力图
plt.figure()
sns.heatmap(corr, annot=True, 
            xticklabels=corr.columns, yticklabels=corr.columns)

plt.show()
```
