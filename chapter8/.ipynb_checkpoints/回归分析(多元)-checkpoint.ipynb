### 产品销量与广告的多元线性回归分析

均方根误差RMSE：预测值与真实值偏差的平方与观测次数n比值的平方根，在sqrt库中

```python
# 调库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics

from copy import deepcopy
import math

plt.rcParams['font.sans-serif'] = ['SimHei']

data = pd.read_csv('Advertising.csv')

# 特征向量
feature_cols = ['TV', 'radio', 'newspaper']
x = data[feature_cols]
y = data['sales']

# 划分训练集和测试集6：4  注意是.4，即0.4，就是40%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.4, random_state=0) # 相当于seed

# 多元回归模型
linreg = LinearRegression()
model = linreg.fit(x_train, y_train)

y_pre = model.predict(x_test)

plt.figure()
plt.plot(range(len(y_pre)), y_pre, 'b', label='预测值')
plt.plot(range(len(y_pre)), y_test, 'r', label='真实值')
plt.legend(loc='upper right')  # 标签右上角显示
plt.xlabel('广告发布量')
plt.ylabel('销售额')
plt.show()
```

### 多元线性回归：numpy实现多元线性回归

```python
x_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453])
```

```python
import numpy as np
import pandas as pd

# 设置随机种子以便于结果复现
np.random.seed(42)

# 生成100个样本的特征数据
n_samples = 100
area = np.random.uniform(50, 200, n_samples)  # 房屋面积（50到200平方米之间）
bedrooms = np.random.randint(1, 6, n_samples)  # 卧室数量（1到5之间）
distance_to_center = np.random.uniform(1, 20, n_samples)  # 距离市中心的距离（1到20公里之间）

# 定义线性方程，添加噪声项 epsilon
epsilon = np.random.normal(0, 10, n_samples)  # 噪声项，标准差为10
price = 0.5 * area + 20 * bedrooms - 0.3 * distance_to_center + 50 + epsilon

# 创建一个包含数据的 DataFrame
data = pd.DataFrame({
    'area': area,
    'bedrooms': bedrooms,
    'distance_to_center': distance_to_center,
    'price': price
})

# 显示生成的数据集
x_train = data[['area', 'bedrooms', 'distance_to_center']].to_numpy()
y_train = data['price'].to_numpy()
```

```python
# 代价函数 
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0.0
    for i in range(m):
        f_wb_i = np.dot(x[i], w) + b  # 预测值 f(w, b) = w*x + b
        cost = cost + (f_wb_i - y[i]) ** 2  # 计算均方误差

    cost = cost / (2 * m)
    return cost

# 梯度
def compute_gradient(x, y, w, b):
    m, n = x.shape  # n行m列，m是样本数量，n是特征数量
    dj_dw = np.zeros((n, ))
    dj_db = 0.

    for i in range(m):
        err = (np.dot(x[i], w) + b) - y[i]  # 计算误差 = 预测值 - 实际值
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err * x[i, j]  # 累加每个特征对应的梯度
        dj_db = dj_db + err

    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_db, dj_dw

# 梯度下降
def gradient_descent(x, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters):
    J_history = []
    w = deepcopy(w_in)
    b = b_in

    for i in range(num_iters):
        dj_db, dj_dw = gradient_function(x, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i < 100000:
            J_history.append(cost_function(x, y, w, b))

        if i % math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")

    return w, b, J_history

# 梯度下降法是通过多次迭代来逐步逼近最优解的，适用于大数据集或计算复杂度较高的情况下。
# 最小二乘法则是通过直接的矩阵求解，一次性得出回归系数的解析解，它在数据量较小或者矩阵运算可以承受的情况下效率更高。
# 最小二乘法
def least_squares(x, y):
    X_b = np.c_[np.ones((x.shape[0], 1)), x]  # 添加常数 1 表示截距项
    theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
    
    b = theta_best[0]  # 截距项
    w = theta_best[1:]  # 回归系数
    return w, b
```

```python
initial_w = np.zeros_like(w_init)
initial_b = 0.
iterations = 1000
alpha = 5.0e-7
w_final, b_final, J_hist = gradient_descent(x_train, y_train, initial_w, initial_b,
                                            compute_cost, compute_gradient,
                                            alpha, iterations)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
m,_ = x_train.shape
for i in range(m):
    print(f"prediction: {np.dot(x_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")

# 拟合曲线
m, _ = x_train.shape
predictions = np.dot(x_train, w_final) + b_final
plt.scatter(range(m), y_train, color='blue', label='Target values')
plt.plot(range(m), predictions, color='red', label='Predictions')
plt.legend()
plt.show()
```
